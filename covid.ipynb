```python
def code(col):
  code=[]
  for i in col:
    try:
      a=pycountry.countries.get(name=i).alpha_3
      code.append(a)
    except:
      code.append('None')
  return code

df3['iso_a3'] = code(df3['Country Name'])
#df3= df3.drop(columns=['code'])
df3.info()
```

```python
from geopy import Nominatim
import geopandas
import numpy as np

map = geopandas.read_file(geopandas.datasets.get_path('naturalearth_lowres'))
map.info()

df4 = pd.merge(map,df3, on='iso_a3')
lat=[]
long=[]
for i in df4['Country Name']:
  try:
    loc = Nominatim(user_agent='ny_explorer')
    loc1 = loc.geocode(i)
    lat.append(loc1.latitude)
    long.append(loc1.longitude)
  except:
    lat.append(np.nan)
    long.append(np.nan)
```
